```python
#importing libraries
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from openpyxl import load_workbook
import mysql.connector
from mysql.connector import Error

df = pd.read_excel (r'C:\Users\sagar\Box\FLEET DATA\salam\uber fleet name & password.xlsx',sheet_name='Sheet1')
df = df[["CITY","Narration name in tally"]]
lst=['6278']
for index, row in df.iterrows():
    if row["CITY"]:
        lst.append(row["Narration name in tally"])
tabs=[]
tabs = ([str(elem) for elem in lst])
print("Tabs contains: ",tabs)

fleet_list=[]
try:
    for i in tabs:
        if i in tabs:
            try:
                sheet_name=""+i+""
                ws='C:\\Users\\sagar\\Box\\FLEET DATA\\salam\\salam all sheets\\'+sheet_name+'.xlsx' #changes
                df = pd.read_excel(ws,sheet_name=0)#changes
                duplicated_columns_list = []
                list_of_all_columns = list(df.columns)
                for column in list_of_all_columns:
                    if list_of_all_columns.count(column) > 1 and not column in duplicated_columns_list:
                        duplicated_columns_list.append(column)
                        duplicated_columns_list
                for column in duplicated_columns_list:
                    list_of_all_columns[list_of_all_columns.index(column)] = column + '_1'
                    list_of_all_columns[list_of_all_columns.index(column)] = column + '_2'
                    list_of_all_columns[list_of_all_columns.index(column)] = column + '_3'

                df.columns = list_of_all_columns
#                 print(df.columns)
                
                
                df_unique=df['Org alias'].unique()
                print("Fleet name :",df_unique,"and code is",i)
                
                #separating gwu columns into two 
                df_not_gwu=df[~df["Description"].str.contains("GwU")]
                df_gwu=df[df["Description"].str.contains("GwU")]
                
                try:
                    df_not_gwu_column=df_not_gwu['Paid to you:Your earnings:Promotion:Promotion'].sum()
                except KeyError:
                    df_not_gwu_column = 0    
#                 print("Paid to you:Your earnings:Promotion:Promotion",df_not_gwu_column)                    
                try:
                    gwu_promotion_column=df_gwu['Paid to you:Your earnings:Promotion:Promotion'].sum()
                except KeyError:
                    gwu_promotion_column = 0    
#                 print("gwu:Paid to you:Your earnings:Promotion:Promotion",gwu_promotion_column)               
                
                #Paid to you:Trip balance:Refunds:Toll and duplicate columns + Paid to you:Trip balance:Expenses:Toll
                try:
                    pty1=df['Paid to you:Trip balance:Refunds:Toll'].sum()
                except KeyError:
                    pty1 = 0    
#                 print("Paid to you:Trip balance:Refunds:Toll",pty1)
                try:
                    pty2=df['Paid to you:Trip balance:Refunds:Toll.1'].sum()
                except KeyError:
                    pty2 = 0    
#                 print("Paid to you:Trip balance:Refunds:Toll.1",pty2)                
                try:
                    pty3=df['Paid to you:Trip balance:Expenses:Toll'].sum()
                except KeyError:
                    pty3 = 0
#                 print("Paid to you:Trip balance:Expenses:Toll",pty3)
                
                #Paid to you:Your earnings:Fare:Adjustment duplicate columns
                try:
                    pty4=df['Paid to you:Your earnings:Fare:Adjustment'].sum()
                except KeyError:
                    pty4 = 0    
#                 print("Paid to you:Your earnings:Fare:Adjustment",pty4)
                
                try:
                    pty5=df['Paid to you:Your earnings:Fare:Adjustment.1'].sum()
                except KeyError:
                    pty5 = 0
#                 print("Paid to you:Your earnings:Fare:Adjustment.1",pty5)                
                
                try:
                    pty6=df['Paid to you:Your earnings:Fare:Adjustment.2'].sum()
                except KeyError:
                    pty6 = 0
#                 print("Paid to you:Your earnings:Fare:Adjustment.2",pty6)                 
                               
                try:
                    pty7=df['Paid to you:Your earnings:Fare:Adjustment.3'].sum()
                except KeyError:
                    pty7 = 0
#                 print("Paid to you:Your earnings:Fare:Adjustment.3",pty7)                                 
                
                columns = pd.read_excel(r'C:\Users\sagar\Box\FLEET DATA\salam\uber fleet name & password.xlsx',sheet_name='columns')
                columns_list=list(columns['columns'])  
                fleet_dict={}
                for col in columns_list:
                    try:
                        pty_columns=df[col].sum()
                    except KeyError:
                        pty_columns = 0
#                     print(col,pty_columns)   
                    fleet_dict.update({
                        "Fleet code":i,
                        "Fleet name":df_unique,
                        "Paid to you:Your earnings:Promotion:Promotion":df_not_gwu_column,
                        "gwu:Paid to you:Your earnings:Promotion:Promotion":gwu_promotion_column,
                        "Paid to you:Trip balance:Refunds:Toll":pty1+pty2+pty3,
                        "Paid to you:Your earnings:Fare:Adjustment":pty4+pty5+pty6+pty7,
                        col:pty_columns})
                fleet_list.append(fleet_dict)
            except:
                print(i,"sheet not found")            
except Error as e:
    print(e,"Some error occured")
    
df_list = pd.DataFrame(fleet_list)
dfm=df_list.T
summ_inp=input("Enter your summary name with date and .csv :- ")
dfm.to_csv(summ_inp,header=None)
print("Summary sheet created : Check the summary sheet from your folder")
```
