```python
#!/usr/bin/env python
# coding: utf-8
# Authot: Sagar Jagnade

import pandas as pd

input_file = "dataset.xlsx"
output_file = "output.xlsx"
output_sheet = "final_output"
allotment_status = "Allotment Status Report"
repair_penalty = "Repair Penalty"
DM_list_history = "DM List History"
DM_type = "DM Type"
final_result = "Final Result"

df_allotment_status = pd.read_excel(input_file, sheet_name=allotment_status, names=['CarNumber', 'Allotment_Date', 'ETM', 'Return_Date'])
df_repair_penalty = pd.read_excel(input_file, sheet_name=repair_penalty, names=['timestamp', 'CarNumber', 'PanelType'])
df_DM_list_hist = pd.read_excel(input_file, sheet_name=DM_list_history, names=['start', 'CarNumber', 'DM', 'end'])
df_DM_type = pd.read_excel(input_file, sheet_name=DM_type)

df_filtered_DM = pd.merge(df_repair_penalty, df_DM_list_hist, how='inner', left_on='CarNumber', right_on='CarNumber').query('timestamp >= start and timestamp <= end')

df_repair_data = df_filtered_DM.merge(df_DM_type, how='inner', left_on='DM', right_on='DM NAME')
df_repair_data.loc[df_repair_data['Cities']=='Mumbai 60:40']
df_repair_data = df_repair_data.drop(['start', 'end', 'DM NAME', 'Cities'], axis=1)

df_repair_data_explode = df_repair_data.assign(PanelType=df_repair_data.PanelType.str.split(',')).explode('PanelType')
df_repair_data_explode[['Panel Type', 'Amount']] = df_repair_data_explode.PanelType.str.split('-', expand=True)
df_repair_data_explode.drop('PanelType', axis=1, inplace=True)
df_repair_data_explode["Panel Type"] = df_repair_data_explode["Panel Type"].str.strip(" ")

df_repair_data_ETM = pd.merge(df_repair_data_explode, df_allotment_status, how='inner', left_on='CarNumber', right_on='CarNumber').query('timestamp >= Allotment_Date and timestamp <= Return_Date')

df_repair_data_ETM.drop(['Allotment_Date','Return_Date'], axis=1, inplace=True)

df_repair_data_ETM.to_excel(output_file, sheet_name=output_sheet,
                            columns=['timestamp', 'CarNumber', 'Panel Type', 'Amount', 'ETM', 'DM'],
                           header=['Timestamp', 'Car Number', 'Panel Type', 'Amount', 'ETM', 'DM Name'],
                           index=False)
```

```python
# Recursive function to generate all permutations of a string
def permutations(remaining, candidate=''):
 
    if len(remaining) == 0:
        print(candidate)
 
    for i in range(len(remaining)):
 
        newCandidate = candidate + remaining[i]
        newRemaining = remaining[0:i] + remaining[i+1:]
 
        permutations(newRemaining, newCandidate)
 
 
if __name__ == '__main__':
 
    s = 'mitsha'
    permutations(s)
```
